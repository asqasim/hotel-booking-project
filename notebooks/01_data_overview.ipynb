```python
import pandas as pd
import numpy as np

try:
    df = pd.read_csv('../data/hotel_bookings.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: File not found. Please ensure 'hotel_bookings.csv' is in the '../data/' directory.")
```

```python
print(f"Shape of dataset: {df.shape}")
print("\nColumns:", df.columns.tolist())
display(df.head())
```

```python
df.info()
```

```python
duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {duplicates}")

# Drop duplicates if any
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Duplicates removed. New shape: {df.shape}")
```

```python
# Unique Values in Categorical Columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n------------------- {col} ----------------------")
    print(df[col].unique()[:10]) # Show first 10 unique values
    print(f"Total Unique: {df[col].nunique()}")
```

```python
df.to_csv('../data/hotel_bookings_clean.csv', index=False)
print("Cleaned dataset saved to '../data/hotel_bookings_clean.csv'")
```
